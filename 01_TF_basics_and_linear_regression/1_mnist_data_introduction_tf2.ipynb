```python
from __future__ import print_function
import tensorflow as tf
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import numpy as np

# (x_train,y_train),(x_test,y_test)=keras.datasets.mnistfashion_mnist.load_data()
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data() # tf2中没有添加mnist数据
```

```python
print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)
print("x_test: ", x_test.shape)
print("y_test: ", y_test.shape)
```

```python
def plot_mnist(data,classes):
    for i in range(10):
        idxs=(classes == i)
        images=data[idxs][0:10]
        
        for j in range(5):
            plt.subplot(5,10,i+j*10+1)
            plt.imshow(images[j].reshape(28,28),cmap='gray')
            if j==0:
                plt.title(i)
            plt.axis('off')
    plt.show()
    
```

```python
# 0代表列，1代表行。
classes=np.argmax(y_train,axis=0) # np.argmax()返回沿轴axis最大值的索引,当axis=1时报错.
plot_mnist(x_train,y_train)
```
